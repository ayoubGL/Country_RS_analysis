# Scoring SUS
SUS yields a single number representing a composite measure of the overall usability of the
system being studied. Note that scores for individual items are not meaningful on their own.
To calculate the SUS score, first sum the score contributions from each item. Each item's
score contribution will range from 0 to 4. For items 1,3,5,7,and 9 the score contribution is the
scale position minus 1. For items 2,4,6,8 and 10, the contribution is 5 minus the scale
position. Multiply the sum of the scores by 2.5 to obtain the overall value of SU.
SUS scores have a range of 0 to 100.

```python
Data_trans = pd.read_csv('usability_survey.csv')
```

```python
Data_trans.head(10)
```

```python
positive_question = Data_trans.loc[[0,2,4,6,8],:]
negative_question = Data_trans.loc[[1,3,5,7,9],:]
```

```python
positive_question = positive_question - 1  # 1 - score
negative_question = 5 - negative_question # 5- score
```

```python
# compute the score of positive question for each user
score_pos = []
for i in positive_question.columns:
    sum_ = 0
    sum_ = sum(positive_question[i])
    score_pos.append(sum_*2.5)
```

```python
# compute the score of negative question for each user
score_neg = []
for i in negative_question.columns:
    sum_ = 0
    sum_ = sum(negative_question[i])
    score_neg.append(sum_*2.5)
```

```python
score_total = []
for i, j in zip(score_neg, score_pos):
    score_total.append(i+j)
```

```python
score_total
import statistics

print(min(score_total))
print(max(score_total))
print(statistics.mean(score_total))
print(statistics.median(score_total))
```

```python
total_Score = sum(score_total)
sus_score = total_Score / 109
sus_score
```
